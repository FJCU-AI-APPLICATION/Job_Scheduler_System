```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 🔹 **參數設定**
num_employees = 10  # 員工總數，可調整
start_date = "2025-03-01"
num_days = 31  # 預測天數，可調整

# 🔹 **生成完整的日期與員工組合**
date_range = pd.date_range(start=start_date, periods=num_days, freq='D')
employee_ids = [f"A{str(i+1).zfill(2)}" for i in range(num_employees)]
df = pd.DataFrame([(date, emp) for date in date_range for emp in employee_ids], columns=["日期", "員工ID"])

# 🔹 **自動生成 `班別`（0=休息, 1=早班, 2=晚班）**
np.random.seed(42)
df["班別"] = np.random.randint(0, 3, size=len(df))

# 🔹 **轉換為橫向顯示（日期為索引，員工ID 為欄位）**
df_pivot = df.pivot(index="日期", columns="員工ID", values="班別")

# 🔹 **轉換數值為班別名稱**
班別對應 = {0: "休息", 1: "早班", 2: "晚班"}
df_pivot = df_pivot.applymap(lambda x: 班別對應[x])

# **輸出結果**
print(df_pivot)
```
